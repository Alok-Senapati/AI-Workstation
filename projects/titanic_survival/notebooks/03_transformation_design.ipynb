# Titanic Survival - Transformation Design

| Feature | Keep / Drop / Create | Transformation | Requires Fitting? | Depends On | Purpose |
|---------|----------------------|----------------|-------------------|------------|---------|
| PassengerId | Drop | None | No | - | Identifier only |
| Survived (Target) | Target | None | No | - | Prediction target |
| Pclass | Keep | None | No | - | Passenger socioeconomic class |
| Name | Drop | Extract Title → Drop | No | - | Raw text not useful after extracting Title |
| Sex | Keep | One-Hot Encoding | Yes | - | Categorical feature |
| Age | Keep | Median Imputation | Yes | - | Handle missing values |
| SibSp | **TBD** | None | No | - | Input to FamilySize (keep/drop to be validated) |
| Parch | **TBD** | None | No | - | Input to FamilySize (keep/drop to be validated) |
| Ticket | Drop | Extract TicketPrefix & TicketGroupSize → Drop | No | - | Raw ticket not useful after feature extraction |
| Fare | Keep | Median Imputation | Yes | - | Handle missing values |
| Cabin | Drop | Extract Deck → Drop | No | - | Raw cabin not useful after extracting Deck |
| Embarked | Keep | Mode Imputation → One-Hot Encoding | Yes | - | Handle missing values and encode categorical feature |
| FamilySize | Create | `SibSp + Parch + 1` | No | SibSp, Parch | Capture family size |
| IsAlone | Create | `FamilySize == 1` | No | FamilySize | Identify solo travelers |
| Title | Create | Extract → Group Rare Categories → One-Hot Encoding | Yes | Name | Capture passenger title / social status |
| Deck | Create | Extract → Missing → "Unknown" → One-Hot Encoding | No* | Cabin | Capture cabin deck information |
| TicketPrefix | Create | Extract → Group Rare Categories → One-Hot Encoding | Yes | Ticket | Capture ticket prefix patterns |
| TicketGroupSize | Create | Count passengers sharing the same ticket | No | Ticket | Capture group travel size (binning TBD) |
